```python
import alpaca
from alpaca import ALPcouplings

import numpy as np
import matplotlib.pyplot as plt

from alpaca.statistics import get_chi2
from alpaca.sectors import default_sectors
from alpaca.plotting.mpl import exclusionplot

plt.rcParams.update({'font.size': 12, 'text.usetex': True, 'font.family': 'serif', 'font.serif': 'Computer Modern Roman'})
```

```python
ma_vec=np.logspace(-2, 1, 100)  # axion mass in GeV
 
fa_vec_1=np.logspace(1,7,100)
fa_vec_2=np.logspace(1,7,100)

fa_1TeV=1e3
couplings=np.logspace(-4, 2, 100)

x_ma_1, y_fa_1=np.meshgrid(ma_vec, fa_vec_1)
x_ma_2, y_fa_2=np.meshgrid(ma_vec, fa_vec_2)

couplings_cq_1= [ALPcouplings({'cqL' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for fa in fa_vec_1]
couplings_cq_2= [ALPcouplings({'cqL' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for c in couplings]
```

```python
couplings_cu_1= [ALPcouplings({'cuR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for fa in fa_vec_1] 
couplings_cu_2= [ALPcouplings({'cuR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', integrator='leadinglog') for c in couplings]
```

```python
couplings_cg_1= [ALPcouplings({'cg' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_1] 
couplings_cg_2= [ALPcouplings({'cg' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
couplings_cd_1= [ALPcouplings({'cdR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_1]
couplings_cd_2= [ALPcouplings({'cdR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
couplings_cl_1= [ALPcouplings({'clL' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_2]
couplings_cl_2= [ALPcouplings({'clL' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
couplings_ce_1= [ALPcouplings({'ceR' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_2]
couplings_ce_2= [ALPcouplings({'ceR' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
couplings_cB_1= [ALPcouplings({'cB' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_2]
couplings_cB_2= [ALPcouplings({'cB' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
couplings_cW_1= [ALPcouplings({'cW' :1.0}, scale= 4*np.pi*fa, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for fa in fa_vec_2]
couplings_cW_2= [ALPcouplings({'cW' :c}, scale= 4*np.pi*fa_1TeV, basis='derivative_above').match_run(10.01, 'VA_below', match_2loops=True) for c in couplings]
```

```python
x_ma, couplings_cu_grid_1 = np.meshgrid(ma_vec, couplings_cu_1)
x_ma, couplings_cu_grid_2 = np.meshgrid(ma_vec, couplings_cu_2)
chi2_cu_1=get_chi2(default_sectors['all'], x_ma, couplings_cu_grid_1, y_fa_1, integrator='no_rge', min_probability=0)
chi2_cu_2=get_chi2(default_sectors['all'], x_ma, couplings_cu_grid_2, fa_1TeV, integrator='no_rge', min_probability=0)
```
