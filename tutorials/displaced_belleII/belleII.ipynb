```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
plt.rcParams.update({'font.size': 12, 'font.family': 'serif', 'font.serif': ['Computer Modern Roman'], 'text.usetex': True})
import os
import re
from alpaca.constants import c_nm_per_ps
```

```python
data_BKee = {}
for file in os.listdir('bkee'):
    if file.endswith('.csv'):
        ctau = float(file[:-4])
        data = pd.read_csv(f'bkee/{file}', names=['m', 'limit'], skiprows=1)
        if len(data) < 1000:
            continue
        data_BKee[ctau] = data
```

```python
matrix_bkee = np.full((len(data_BKee), len(data_BKee[0.01])), np.nan)
for i, ctau in enumerate(sorted(data_BKee.keys())):
    data = data_BKee[ctau]['limit'].values
    matrix_bkee[i,:len(data)] = data
```

```python
x_ma, y_ctau = np.meshgrid(data_BKee[0.01]['m'].values, sorted(data_BKee.keys()))
```

```python
plt.pcolor(x_ma, np.log10(y_ctau*1e7/c_nm_per_ps), np.log10(matrix_bkee), cmap = 'jet_r', vmin=-7.5, vmax=-1.5)
ax = plt.gca()
ax.set_yticks([-1, 0, 1, 2, 3])
ax.set_yticklabels([r'$10^{-1}$', r'$1$', r'$10$', r'$10^{2}$', r'$10^{3}$'])
cb = plt.colorbar()
cb.set_ticks([-7, -6, -5, -4, -3, -2])
cb.set_ticklabels([r'$10^{-7}$', r'$10^{-6}$', r'$10^{-5}$', r'$10^{-4}$', r'$10^{-3}$', r'$10^{-2}$'])
plt.xlabel(r'$m_{a}$ [GeV]')
plt.ylabel(r'$\tau_a$ [ps]')
cb.set_label(r'$\mathrm{BR}(B^+ \to K^+ a)\times\mathrm{BR}(a \to e^+ e^-)$')
```

```python
matrix_res = np.full((matrix_bkee.shape[0]+1, matrix_bkee.shape[1]+1), np.nan)
matrix_res[:-1,:-1] = np.log10(matrix_bkee)
matrix_res[-1,:-1] = data_BKee[0.01]['m'].values
matrix_res[:-1,-1] = np.log10(np.array(sorted(data_BKee.keys()))*1e7/c_nm_per_ps)
```

```python
np.save('../../alpaca/experimental_data/visible/belleII_BKee_displ.npy', matrix_res)
```
